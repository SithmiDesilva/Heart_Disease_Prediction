```python
# Plot box plots for specific numerical variables to identify outliers
for column in ['age', 'sex','cp','trestbps','chol','tbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']:
    if column in heart_data.select_dtypes(include=['number']).columns:
        plt.figure(figsize=(7, 5))
        sns.boxplot(data=heart_data, x=column, color='lightcoral')
        plt.title(f"Box Plot of {column}")
        plt.xlabel(column)
        plt.grid(axis='y', alpha=0.75)
        plt.show()
```

```python


# Assuming your data is in a pandas DataFrame called 'heart_data'
# and the columns with outliers are 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'thal'

columns_with_outliers = ['trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'thal']

# Calculate the IQR for each column
Q1 = heart_data[columns_with_outliers].quantile(0.25)
Q3 = heart_data[columns_with_outliers].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace outliers with the respective bounds (optional)
# This method is called "Winsorization"
for column in columns_with_outliers:
    heart_data[column] = np.clip(heart_data[column], lower_bound[column], upper_bound[column])

# Filter the data to exclude outliers (optional)
# This method removes entire rows containing outliers
filtered_data = heart_data[
    (heart_data[columns_with_outliers] >= lower_bound) & (heart_data[columns_with_outliers] <= upper_bound)
].all(axis=1)
heart_data = heart_data[filtered_data]

# Optionally, you can replace outliers with NaN values and impute them using various techniques
# heart_data[column] = np.where(heart_data[column] < lower_bound[column], np.nan, heart_data[column])
# heart_data[column] = np.where(heart_data[column] > upper_bound[column], np.nan, heart_data[column])
# heart_data.fillna(heart_data.mean(), inplace=True)  # Or use other imputation strategies
```
